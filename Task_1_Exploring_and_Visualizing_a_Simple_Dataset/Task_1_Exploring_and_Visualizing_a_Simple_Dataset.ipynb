### Objective: Load, inspect, and visualize the Iris dataset to understand its structure and patterns.
### Libraries Used: pandas, matplotlib, scikit-learn

```python
# 1. IMPORT REQUIRED LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
# 2. LOAD AND PREPARE THE DATASET
iris=load_iris()
```

```python
iris
```

Loads the Iris dataset from scikit-learn and converts it into a pandas DataFrame with human-readable species names.
  Returns: pd.DataFrame: Clean DataFrame with 4 features + 1 'species' column.
    

```python
i_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
i_df['species']=iris.target
```

```python
# Map numeric labels (0,1,2) to actual species names for clarity
species_map={0:'setosa',1:'versicolor',2:'virginica'}
i_df['species']=i_df['species'].map(species_map)
```

The Iris dataset has **150 samples**—meaning measurements from 150 flowers, evenly split across three species (50 each).

Looking at the **average (mean)** values:  
- Sepal length is about **5.84 cm**  
- Sepal width is around **3.06 cm**  
- Petal length is roughly **3.76 cm**  
- Petal width averages **1.20 cm**

The **standard deviation** tells us how spread out the values are:  
- Petal length and width vary a lot more than sepal width. For example, petal length has a high spread (**±1.77 cm**), while sepal width is pretty consistent (**±0.44 cm**).

From the **min and max** values:  
- The shortest sepal is **4.3 cm**, the longest is **7.9 cm**  
- The narrowest petal is just **0.1 cm** wide, while the widest is **2.5 cm**—that’s a big difference!

The **median (50%)** gives a better sense of the “middle” value, especially when data isn’t perfectly symmetrical:  
- Half the flowers have a petal length **below 4.35 cm**, and half above.  
- For sepal width, the median is **3.0 cm**, very close to the mean—so it’s fairly symmetric.

Also notice:  
- The **interquartile range** (from 25% to 75%) for petal measurements is much wider than for sepal width. That means petal sizes differ a lot more between species, which is why they’re super useful for telling the flowers apart.



### 4. CREATE VISUALIZATIONS


    Generates three types of plots:
    1. Scatter plot: Sepal Length vs Sepal Width (colored by species)
    2. Histograms: Distribution of all numeric features
    3. Box plots: Detect outliers in each feature


```python
# 1. Scatter plot: Sepal Length vs Sepal Width (colored by species
plt.figure(figsize=(8,6))
sns.scatterplot(data=i_df,x=i_df['sepal length (cm)'],y=i_df['sepal width (cm)'],hue=i_df['species'])
plt.title('Scatter plot: Sepal length VS Sepal width')
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Sepal width(cm)')
plt.grid(True)
plt.legend()
plt.show()
```


This is a **scatter plot comparing sepal length (x-axis) vs. sepal width (y-axis)** for 150 Iris flowers, grouped by species:  
- **Blue dots = setosa**  
- **Orange dots = versicolor**  
- **Green dots = virginica**

1. **Setosa (blue)** is clearly separated from the other two:
- Has the **shortest sepal length** (mostly under 5.5 cm)
- Has **wider sepals** (mostly above 3.0 cm)
- Forms a tight, distinct cluster in the bottom-left

*This means: If you see a flower with a short sepal and wide sepal, it’s almost certainly setosa.*

2. **Versicolor (orange)** and **Virginica (green)** overlap a lot:
- Both have **longer sepals** (mostly 5.5 cm and up)
- Virginica tends to have **slightly longer sepals** than versicolor
- Versicolor has **narrower sepals** on average (mostly 2.5–3.0 cm), while virginica is more spread out (2.5–3.8 cm)

*This means: It’s harder to tell versicolor and virginica apart just by sepal size alone — you’d need petal measurements too.*

3. **No strong linear relationship between sepal length and width**:
- The points don’t form a clear diagonal line → so **sepal length doesn’t predict sepal width well**.
- Within each species, there’s still variation — but the overall pattern per species is consistent.

In Simple Terms:

“If you only look at sepal size, you can easily spot setosa flowers — they’re small and wide. But versicolor and virginica look too similar in sepal size to tell apart reliably.”


```python
# 2.HISTOGRAMS: Distribution of each numerical features
numerical_feature=['sepal length (cm)','sepal width (cm)', 'petal length (cm)','petal width (cm)']
num_data=i_df[numerical_feature]
```

```python
plt.figure(figsize=(10,7))
num_data.hist(bins=15,color='lightcoral',edgecolor='black')
plt.suptitle('Distribution of Numerical Features')
plt.tight_layout()
plt.grid(True,axis='y')
plt.show()
```

Histograms:

Sepal length: Mostly between 5–7 cm, kinda bell-shaped.

Sepal width: Peaks around 3.0 cm, slightly skewed left.

Petal length: Bimodal — big peak near 1.5 cm (setosa), smaller one near 4–5 cm (versicolor/virginica).

Petal width: Same story — most under 0.5 cm (setosa), then another group around 1.0–2.0 cm.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=i_df.drop('species', axis=1))
plt.title('Box Plots (Seaborn)')
plt.grid(True)
plt.show()
```

Box plots:

Shows spread + outliers.

Sepal length/width: Tight boxes → less variation.

Petal length/width: Wider boxes + more outliers → way more variation between species.

Petals are way better for telling species apart than sepals.
